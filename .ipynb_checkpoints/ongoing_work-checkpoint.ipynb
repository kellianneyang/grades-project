```python
# instantiate pca
pca = PCA(n_components = 464)

# fit pca on preprocessed_and_scaled_df_fe
pca.fit(preprocessed_and_scaled_df_fe)
```

```python
len(pca.explained_variance_ratio_)
```

```python
# let's visualize the cummulative sum of the variances 

plt.style.use("seaborn")
plt.figure(figsize=(150,8))
exp_var = pd.Series(pca.explained_variance_ratio_)

# Plot the variance for each component
exp_var.plot(kind="bar", alpha=0.7) # plot bar chart

# Calculate the amount of variance explained added by each additional component
total = 0
var_ls = []
for x in exp_var:
    total = total+x
    var_ls.append(total)
    
# Plot line chart of increasing variances
pd.Series(var_ls).plot(alpha=0.7) 
# marker="o"

# Plot labels and titles
plt.xlabel("Principle Components", fontsize="x-large")
plt.ylabel("Percentage Variance Explained", fontsize="x-large")
plt.title("grades Scree Plot", fontsize="xx-large")

# 80 and 90 percent lines
plt.axhline(y = 0.8, color = 'r')
plt.axhline(y = 0.9, color = 'r')
plt.show();
```

To retain 80% of our data, we could cut down the number of PCs to 276. From 464, that represents a reduction of about 40%. 

To retain more data, say 90%, we could cut down to 329 PCs, a reduction of about 30%. 

For final tuning, let's work with 276 components to find the best hyperparameters for our model faster.
